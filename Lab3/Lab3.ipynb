```python
import csv
```

```python
f = open('cig.csv', 'rb')
reader = csv.reader(f)
Tar_content = []
Nicotine_content = []
for row in reader:
    b = row[0].replace("\t" , " ")
    c = b.split()
    d = [c[0]]
    e = map(float , c[1:])
    d.extend(e)
    Tar_content.append(e[0])
    Nicotine_content.append(e[1])
    print d
    
    
    
    
f.close()
```
